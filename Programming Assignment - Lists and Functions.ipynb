```python
# --- Task 7.4 ---
# Make a list called things with these three strings as elements
things = ["mozzarella", "cinderella", "salmonella"]
print("7.4 List:", things)

# --- Task 7.5 ---
# Capitalize the element in things that refers to a person and then print the list.
# The person is "cinderella" at index 1.
print("\n7.5 Capitalized element:", things[1].capitalize())
print("7.5 List after printing:", things)
# ANSWER: Did it change the element in the list?
# No. The .capitalize() method returns a new string but does not modify the original list.

# --- Task 7.6 ---
# Make the cheesy element of things all uppercase and then print the list.
# The cheesy element is "mozzarella" at index 0.
things[0] = things[0].upper()
print("\n7.6 List after uppercase change:", things)

# --- Task 7.7 ---
# Delete the disease element from things and print the list.
# The disease element is "salmonella" at index 2.
del things[2]
# Alternatively, you could use: things.remove("salmonella")
print("\n7.7 List after deletion:", things)


# --- Task 9.1 ---
# Define a function called good() that returns the following list:
# ['Harry', 'Ron', 'Hermione']
def good():
    return ['Harry', 'Ron', 'Hermione']

# Call the function to verify it works (optional, but good practice)
print("9.1 Output:", good())


# --- Task 9.2 ---
# Define a generator function called get_odds() that returns the odd numbers from range(10).
def get_odds():
    for number in range(1, 10, 2):
        yield number

# Use a for loop to find and print the third value returned.
count = 0
for number in get_odds():
    count += 1
    if count == 3:
        print("9.2 Third odd number is:", number)
        break
```

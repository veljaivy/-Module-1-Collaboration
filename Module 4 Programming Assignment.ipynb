```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///books.db')
```

```python
from sqlalchemy.orm import sessionmaker, declarative_base
from sqlalchemy import Column, String, Integer

Base = declarative_base()

class Book(Base):
    __tablename__ = 'book' # Match the table name in your database
    title = Column(String, primary_key=True) # Assuming title is unique or primary key for simplicity
    author = Column(String) # Add other columns as they exist in your table
    year = Column(Integer)

# Create the tables in the database (if they don't exist)
Base.metadata.create_all(engine)

# Create a session to interact with the database
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# Query the title column and order the results alphabetically by title
results = session.query(Book.title).order_by(Book.title).all()

# Print the titles
for title in results:
    print(title)  # Changed from print(title[0])
```

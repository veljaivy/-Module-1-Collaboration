## 4.1 Secret Number
Choose a number between 1 and 10 for `secret` and `guess`, then run conditional tests.

```python
secret = 7
guess = 4

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

## 4.2 Small and Green
Assign `True` or `False` to `small` and `green` to find a match among cherry, pea, watermelon, and pumpkin.

```python
small = True
green = False

if small:
    if green:
        print("pea")
    else:
        print("cherry")
else:
    if green:
        print("watermelon")
    else:
        print("pumpkin")
```

## 6.2 While Loop Comparison
Use a `while` loop to compare `number` with `guess_me`.

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

## 6.3 For Loop Range
Use a `for` loop to iterate over `range(10)` to find `guess_me`.

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
